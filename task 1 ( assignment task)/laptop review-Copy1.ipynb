```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings("ignore")

# Set plot styles
sns.set(style="whitegrid")
```

```python
# Load the dataset (adjust filename if different)
df = pd.read_csv('flipkart_laptop_reviews.csv')

# Display the first 5 rows
df.head()
```

```python
# Check the shape of the dataset
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns")

# Get a summary of the dataset
df.info()

# Check for missing values
df.isnull().sum()
```

```python
import numpy as np

# Remove commas and convert to int
df['no_ratings'] = df['no_ratings'].str.replace(',', '').astype(int)
df['no_reviews'] = df['no_reviews'].str.replace(',', '').astype(int)

# Extract brand from product_name
df['brand'] = df['product_name'].str.split().str[0]

# Create sentiment category
def get_sentiment(score):
    if score >= 4:
        return 'Positive'
    elif score == 3:
        return 'Neutral'
    else:
        return 'Negative'

df['sentiment'] = df['rating'].apply(get_sentiment)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Rating distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='rating', palette='viridis')
plt.title("Distribution of User Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Top brands by count
top_brands = df['brand'].value_counts().nlargest(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='magma')
plt.title("Top 10 Laptop Brands by Number of Reviews")
plt.ylabel("Number of Reviews")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.show()

# Average rating by brand (top 10 brands only)
avg_rating = df.groupby('brand')['rating'].mean().sort_values(ascending=False)
avg_rating_top10 = avg_rating.loc[top_brands.index]

plt.figure(figsize=(8, 4))
sns.barplot(x=avg_rating_top10.index, y=avg_rating_top10.values, palette='coolwarm')
plt.title("Average Rating by Top Brands")
plt.ylabel("Average Rating")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count sentiments
sentiment_counts = df['sentiment'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='pastel')
plt.title("Sentiment Distribution")
plt.ylabel("Number of Reviews")
plt.xlabel("Sentiment")
plt.show()
```

```python
# Create column for SSD presence
df['has_ssd'] = df['product_name'].str.contains("SSD", case=False)

# Compare average ratings
ssd_rating = df.groupby('has_ssd')['rating'].mean()

plt.figure(figsize=(6, 4))
sns.barplot(x=ssd_rating.index.map({True: 'Has SSD', False: 'No SSD'}), y=ssd_rating.values, palette='Set2')
plt.title("Average Rating: SSD vs Non-SSD")
plt.ylabel("Average Rating")
plt.xlabel("SSD Presence")
plt.ylim(0, 5)
plt.show()
```
